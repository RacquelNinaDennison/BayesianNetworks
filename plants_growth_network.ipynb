Plant growth classification Bayesian Network 

```python
from pylab import * 
import matplotlib.pyplot as plt
import os
import pyAgrum as gum 
import pyAgrum.lib.notebook as gnb
```

```python
# setting up the network 
bn = gum.BayesNet("PlantGrowthDataSet")
id_h=bn.add(gum.LabelizedVariable('h','humidity',2))
id_t = bn.add(gum.LabelizedVariable('t','tempreture',2))
id_f = bn.add(gum.LabelizedVariable('f','fertilizer',2))
id_w = bn.add(gum.LabelizedVariable('w','water',2))
id_s = bn.add(gum.LabelizedVariable('s','soil',2))
id_g = bn.add(gum.LabelizedVariable('g','growth',2))

for link in [(id_h,id_s),(id_t,id_s), (id_f,id_s), (id_f, id_g), (id_w, id_g), (id_s, id_g)]:
    bn.addArc(*link)
bn
```
